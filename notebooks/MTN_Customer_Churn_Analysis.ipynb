# MTN Customer Churn Analysis

## 1. Problem Statement
The main business problem is:  
**"Why are some MTN customers leaving (churning), and how can we predict, prevent, or reduce churn?"**  

Understanding customer churn will help MTN improve retention, increase revenue, and optimize customer engagement.

---

## 2. Key Questions

### 2.1 Descriptive Questions
- What is the overall churn rate?
- Which states or regions have the highest churn?
- Which subscription plans or devices are most associated with churn? Which ones have the highest retention?
- How does churn differ by gender or age group? (Consider age ranges: 18–25, 26–35, etc.)
- What is the average tenure of churned vs. retained customers?

### 2.2 Diagnostic Questions
- Are low satisfaction rates driving churn?
- Do high or low data usage customers churn more? (Consider splitting usage into tiers: low, medium, high)
- Are frequent buyers less likely to churn?
- Do certain price points or subscription plans correlate with higher churn? (Include bundled offers if applicable)
- What are the most common reasons for churn? (Categorize reasons: service issues, cost, competitor offers, etc.)

### 2.3 Predictive Questions
- Can we predict whether a customer will churn next month based on past behavior?
- Which combination of factors (tenure, satisfaction, plan, usage, revenue, device type) are strongest churn predictors?

### 2.4 Prescriptive Questions
- Which customers should we target with retention campaigns? (Use priority scoring: high, medium, low)
- Which subscription plans or offers could reduce churn? (Evaluate discount vs. no discount scenarios)
- How can MTN improve customer satisfaction to retain more customers? (Focus on support, billing, network experience, etc.)
