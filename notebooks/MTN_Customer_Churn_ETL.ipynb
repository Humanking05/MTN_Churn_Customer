```python
# Create a database

# Connect to SQLite (creates database if it doesn't exist)
conn = sqlite3.connect("mtn_churn.db")
cursor = conn.cursor()
```

```python
# Load dataframe into SQLite
df.to_sql("customers", conn, if_exists="replace", index=False)
```

```python
# Check if data was loaded
pd.read_sql("SELECT * FROM customers LIMIT 5;", conn)
```

- You can now query the `customers` table for analysis.
